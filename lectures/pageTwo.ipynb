# Ambrosia Transformations

#### With ambrosia, you can do different transformations on the objects and shapes you create. As you have probably noticed from the previous page, most objects created and displayed were mostly centered on the image and roughly about the same size. However, you will realize quickly that this will need to change when creating more complex objects. This is because you may need to shrink the size of some objects and put objects on top of one another inorder to construct other objects which is why transformations are so powerful. 

#### By default, shapes and objects created are centered at the origin of an 3 dimensional plane(x, y, z). The following are the dimensions of ambrosia's predefined shapes in units: 
* cube -> 100 x 100 x 100
* sphere -> diameter of 100
* cylinder -> height and diameter of 100 and oriented along the y-axis
* cone -> height and diameter of 100 and pointing upward along the y-axis

#### We can do the following transformations on objects created in ambrosia:
* scale -> shrink or enlarge objects in all dimensions toward or away from the origin
* xRot, yRot, zRot -> rotate an object about an axis
* translate -> move an object relative to its current position

#### Consider the following yellow plastered cube from the previous page

```python
def reset():
    global sphere, cube, cylinder, cone, scene, camera, image, bulb
    sphere = Sphere()
    cube = Cube()
    cylinder = Cylinder()
    cone = Cone()
    bulb = Light().color(white)
    scene = Group().add(bulb,translate(0,300,-300))
    camera = Camera().subject(scene)
    image = camera.getImage()

from ambrosia import *
reset()

scene.add(cube, yellowPlaster)

camera.shoot(scene)
```

#### Let us shrink it to about half it's size. Notice that we are using 0.5 because we want ambrosia to make the cube half = 0.5 its original size. We can expand to twice its original size by giving scale the value 2 instead of 0.5 because we would be telling ambrosia to do just that with the cube we created.

```python
from ambrosia import *
reset()

scene.add(cube, yellowPlaster, scale(0.5))

camera.shoot(scene)
```

#### We can also tell ambrosia to scale in a specific direction (either x, y or z). 

#### For example, if we wanted to shrink the size of the cube to half in the x direction but leave the other directions the same, would give scale three values, the first (0.5) representing the size of cube in the x direction, the next (1) representing the size of the cube in y direction and the last (1) representing the size of the cube in the z direction as shown below:

```python
from ambrosia import *
reset()

scene.add(cube, yellowPlaster, scale(0.5, 1, 1))

camera.shoot(scene)
```

#### The next type of transformation available in ambrosia is rotations. You can rotate an object an axis. This is very helpful as it enables you to see different parts of an object you create. As explained earlier, you can rotate an object along all three dimensional axes using xRot, yRot and zRot. Ambrosia rotations use degrees as the measurement of the rotation angle.

#### Consider the following black plastered cone rotated 30 degrees along the x-axis

```python
from ambrosia import *
reset()

scene.add(cone, blackPlaster,xRot(30))

camera.shoot(scene)
```

#### Consider the same cone now rotated 30 degrees along all axes.

```python
from ambrosia import *
reset()

scene.add(cone, blackPlaster,xRot(30)*yRot(30)*zRot(30))

camera.shoot(scene)
```

#### The last type of transformation is translation. This moves an object relative to its current position. This is very helpful in placing objects at specific positions next or relative to other objects. Translate takes three values (x movement, y movement and z movement). Consider the following gray plastered sphere moved 20 units to the right of its original position. Notice that we are giving 0 values to y and z movements because we only want to move the sphere in the x direction. Positive values indicate up and to the right movements while negative values indicate down and to the left movements.

```python
from ambrosia import *
reset()

scene.add(sphere, grayPlaster,translate(20,0,0))

camera.shoot(scene)
```

As you progress along, you will need to use multiple transformations on an object to attain a desired result. Ambrosia enables you to do this by making it possible to apply multiple transformations on an object at the same time. You will do more of this later, but for now, check out this example where multiple transformations are done on a purple cube

```python
from ambrosia import *
reset()

scene.add(cube, purplePlaster,translate(0,50,0)*scale(0.4,1,0.4)*scale(0.5)*xRot(-90)*yRot(30))

camera.shoot(scene)
```
