# A Better Bracelet
A demonstration of lofting.

```python
from ambrosia import *

# the material we love
gold=Material().color((0.8,0.7,0))
gold.diffuse(1).reflection(0.3).specularity(1).roughness(0.05)

# number of sides in cross section
n=3
cross = raisePoly(polygon(n))
cross = scale(.5)(cross)  # scale cross section

# build scaffolding:
l = []
for angle in range(0,360+1,1):
    t = zRot(angle/n)*translate(100,0,0)*yRot(angle)
    sect = t(cross) # t.mapPoly(cross)
    l.append(sect) # l = l + [sect]

# loft over the scaffolding
sculpt = loft(*l)

scene.add(sculpt,xRot(-60),gold)
image.background( (0.75,0.75,1) ).antiAlias(1)

camera.shoot()
```
